# ğŸ§  Sequential (Linear) Prompt Chaining â€” Concept and Working Example (Cloud Version)

**Definition:**  
Sequential (or Linear) Prompt Chaining means connecting multiple AI prompts so that:

> **Each stepâ€™s output becomes the input for the next prompt.**

Itâ€™s like a **reasoning pipeline**, where the AI builds on its own earlier results step by step.

---
### ğŸ§© Example of Flow
```
Prompt 1 â†’ Output A  
Prompt 2 (uses A) â†’ Output B  
Prompt 3 (uses B) â†’ Output C
```
Each step refines or extends the previous one.

---

## ğŸ’» Full Code Implementation (with Explanations)

Below is a complete working example of Sequential Prompt Chaining using **Ollama Cloud** and the `gpt-oss:120b` model.

Each step logically depends on the previous output, demonstrating a **linear reasoning flow**.

````python
# ============================================================
# ğŸ§  Sequential (Linear) Prompt Chaining â€” Cloud Version
# ============================================================

# Import required libraries
import os
from dotenv import load_dotenv
from ollama import Client

# Step 1ï¸âƒ£ â€” Load environment variables from .env file
load_dotenv()

# Retrieve API key securely
api_key = os.getenv('OLLAMA_API_KEY')
if not api_key:
    raise ValueError('âš ï¸ OLLAMA_API_KEY not found. Check your .env file path or variable name.')

# Step 2ï¸âƒ£ â€” Initialize the Ollama Cloud client
client = Client(
    host='https://ollama.com',
    headers={'Authorization': f'Bearer {api_key}'}
)

# Define helper function for prompt chaining
def ask(prompt: str) -> str:
    messages = [{'role': 'user', 'content': prompt}]
    full_text = ''

    # Stream the chat output from Ollama Cloud
    for part in client.chat('gpt-oss:120b', messages=messages, stream=True):
        content = part['message']['content']
        print(content, end='', flush=True)
        full_text += content

    print('\n')
    return full_text.strip()

# ============================================================
# ğŸªœ Step 1 â€” Pick a Business Area
# ============================================================
step1 = ask('Pick a business area that might be worth exploring for an Agentic AI opportunity.')
print('Step 1 â€” Business Area:\n', step1, '\n')

# ============================================================
# ğŸ”¹ Step 2 â€” Identify a Pain Point
# ============================================================
step2_prompt = f'Within {step1}, describe a key pain-point or challenge that may benefit from Agentic AI.'
step2 = ask(step2_prompt)
print('Step 2 â€” Pain Point:\n', step2, '\n')

# ============================================================
# ğŸ”¸ Step 3 â€” Propose an Agentic AI Solution
# ============================================================
step3_prompt = f'Propose an Agentic AI solution to address this pain-point: {step2}'
step3 = ask(step3_prompt)
print('Step 3 â€” Agentic AI Solution:\n', step3)
````

## ğŸ§  Explanation of the Workflow

| Step | Purpose | Input | Output Example |
|------|----------|--------|----------------|
| **Step 1** | Identify a business area | Static prompt | â€œHealthcareâ€ |
| **Step 2** | Describe a pain point in that area | Output from Step 1 | â€œPatient triage inefficiencyâ€ |
| **Step 3** | Propose AI solution | Output from Step 2 | â€œAI-powered triage scheduling assistantâ€ |
