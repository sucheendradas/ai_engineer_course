Loading Api

```python
import os
from dotenv import load_dotenv
load_dotenv()
  
from sarvamai import SarvamAI


```

Translate

```python
client = SarvamAI(
    api_subscription_key=os.environ["SARVAM_API_KEY"],
)

response = client.text.translate(
    input="Hi, My Name is Vinayak. who is the prime minister of India?",
    source_language_code="auto",
    target_language_code="ml-IN",
    speaker_gender="Male"
)

print(response)
```

Chat Completion

```python
#LLM chat formal free 

response = client.chat.completions(messages=[
    {"role": "user", "content": "Hey, what is the capital of India?"}])
print(response)
```

```python
mytext = """
ഞാനൊരു മലയാളി എന്നും മണ്ണിൻ കൂട്ടാളി
എങ്ങും അതിരുകളില്ല മതിലുകളില്ലാ സ്നേഹത്തേരാളി
മണ്ണാണ് ജീവൻ മണ്ണിലാണ് ജീവൻ
പൊന്നിൻ വിളകൾ കൊയ്തെടുത്തൊരു സ്വർഗ്ഗം തീർക്കും ഞാൻ
ഇവിടൊരു സ്വർഗ്ഗം തീർക്കും ഞാൻ
"""
```

```python

from sarvamai.play import play, save

response = client.text_to_speech.convert(
    text=mytext,
    target_language_code="ml-IN",
    pace = .75,
    speaker="karun",
    enable_preprocessing=True,
)

#play(response)
save(response, "output.wav")
```
